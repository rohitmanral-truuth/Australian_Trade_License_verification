```python
# Python program to convert a list to string
    
# Function to convert  
def listToString(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for ele in s: 
        str1 += ele  
    
    # return string  
    return str1 
```

```python
import selenium
import time
from selenium import webdriver

from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
name=[]
status=[]
address=[]
lic_by=[]
lic_class=[]
start_date=[]
exp_date=[]
    
    
for i in range(len(licence_number_list)):
    driver.get('https://www.licensedtrades.com.au/')
    time.sleep(1)
    
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[3]/div/form/div[1]/div[2]/input').send_keys(licence_number_list[i])
    driver.find_element_by_xpath('//*[@id="advanced_trigger"]').click()
    time.sleep(1)
    
    driver.find_element_by_xpath('//*[@id="advanced_section"]/div/div[3]/p[6]/input').click()
    driver.find_element_by_xpath('//*[@id="advanced_section"]/div/div[3]/p[2]/input').click()
    time.sleep(2)
    time.sleep(1)
    
    driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[3]/div/form/div[1]/div[3]/input').click()
    time.sleep(1)
        
    driver.find_element_by_xpath('//*[@id="search_results_licence"]/div/div/div[1]/div/div[1]/div/p/a').click()
    time.sleep(1)
        
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')    
        
    name.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[1]/div/h3").text) #Business name
    
    status.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[1]/div/p/span").text) #Lic Status
    
    address.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[2]/p[1]").text) #address
    
    lic_by.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[2]/p[3]").text)
             
    try:
        licensed_class_b = [i.text for i in soup.findAll('td', {'class': 'bbssghl brssghl pls'})]
        licensed_class_b = [x for x in licensed_class_b if not isinstance(x, int)]
        for x in licensed_class_b:
            if x == ' - ':
                licensed_class_b.remove(x)
        converted_list = []
        for element in licensed_class_b:
            converted_list.append(element.strip())
        lic_class.append(converted_list)
    except:
        lic_class.append("-")
        continue
                                  
    try:
        start_date.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[4]/table/tbody/tr/td[2]").text)
    except:
        start_date.append("-")
                                  
    try:
        exp_date.append(driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[4]/div/div[1]/div[4]/table/tbody/tr/td[3]").text)
    except:
        exp_date.append("-")
                                  

       
new_a=[]
for i in address:
    new_a.append((i.split(":"))[1].strip())    
    
    
new_lic_by=[]
for i in lic_by:
    new_lic_by.append((i.split(":"))[1].strip())     
            
            
print(licence_number_list)
print(name)
print(status)
print(new_a)
print(new_lic_by)
print(lic_class)
print(start_date)
print(exp_date)
```

```python
# initialise data of lists.
data = {'Licence number':licence_number_list, 'Name': name, 'Status':status, 'Address':new_a, 'Licensed By': new_lic_by, 'License Class':lic_class, 'Start date':start_date, 'Expiry date':exp_date}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
df
```
